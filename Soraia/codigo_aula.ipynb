```python

nome_coluna = 'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)'

#fórmula q tá no caderno 
df_coluna_normalizada = (df[nome_coluna] - df[nome_coluna].min()) / (df[nome_coluna].max() - df[nome_coluna].min()) 

#depois de normalizar

print(df_coluna_normalizada.mean()) 
print(df_coluna_normalizada.std())

#antes de normalizar

print(df[nome_coluna].mean())
print(df[nome_coluna].std())
```

```python
#função para normalizar os dados
def normalizar(nome_coluna):
  return (df[nome_coluna] - df[nome_coluna].min()) / (df[nome_coluna].max() - df[nome_coluna].min())


print(df['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)'].mean())
print(df['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)'].std())
df_coluna_normalizada = normalizar('PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)')
print(df_coluna_normalizada.mean())
print(df_coluna_normalizada.std())

print()

print(df['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'].mean())
print(df['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'].std())
df_coluna_normalizada = normalizar('TEMPERATURA DO AR - BULBO SECO. HORARIA (C)')
print(df_coluna_normalizada.mean())
print(df_coluna_normalizada.std())
```
